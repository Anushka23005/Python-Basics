12. **How does the elif statement work ?**
- The elif is used when you need to check **multiple conditons**.It comes after if and before else.

- If the first if condition is True → its block runs, and the rest are skipped.

- If if is False, Python checks the elif condition.

- If elif is True then that block runs, and the rest are skipped.

- If none are True, the else block runs.


```python
#Example of elif satement
marks = 75

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

```python
#6. Write a function to demonstrate the behavior of mutable and immutable arguments.
def demo_mutable_immutable(a, b, c):
    # a is an integer (immutable)
    # b is a list (mutable)
    # c is a string (immutable)

    print("Initial values:")
    print(f"a = {a}, b = {b}, c = {c}")

    # Trying to modify the arguments
    a += 1
    b.append(4)
    c += " World"

    print("\nAfter modification:")
    print(f"a = {a}, b = {b}, c = {c}")

# Testing the function
a = 1
b = [1, 2, 3]
c = "Hello"

print("Before function call:")
print(f"a = {a}, b = {b}, c = {c}")

demo_mutable_immutable(a, b, c)

print("\nAfter function call:")
print(f"a = {a}, b = {b}, c = {c}")
```

```python
#14. Write a program to reverse a string using a while loop.
word = "Hell0"

i = len(word) - 1
reversed_str = ""

while i >= 0:
    reversed_str += word[i]
    i -= 1

print("Original String:", word)
print("Reversed String:", reversed_str)
```
